```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['gamers'], bins=10, kde=True, color='skyblue')
plt.title('DistribuciÃ³n del NÃºmero de Jugadores')
plt.xlabel('Jugadores')
plt.ylabel('Frecuencia')
```

```python
plt.subplot(1, 3, 2)
sns.histplot(df['skins'], bins=10, kde=True, color='salmon')
plt.title('DistribuciÃ³n de las Apuestas (Items)')
plt.xlabel('Items')
plt.ylabel('Frecuencia')
```

```python
plt.subplot(1, 3, 3)
sns.histplot(df['money'], bins=10, kde=True, color='lightgreen')
plt.title('DistribuciÃ³n del Dinero Movido')
plt.xlabel('Dinero ($)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df['hour'] = df['time'].dt.hour
```

```python
df['day'] = df['time'].dt.date
```

```python
df['time_hhmm'] = df['time'].dt.strftime('%H:%M')
```

```python
df.head()
```

```python
from sqlalchemy import create_engine, text
# motor SQLAlchemy
engine = create_engine('mysql+mysqlconnector://root:ContraseÃ±a.@localhost/prueba')

nombre_tabla = 'casino'

# Eliminamos la tabla si ya existe
with engine.connect() as connection:
    drop_table_statement = text(f"DROP TABLE IF EXISTS {nombre_tabla}")
    connection.execute(drop_table_statement)

df.to_sql(nombre_tabla, con=engine, if_exists='replace', index=False)

print(f"La tabla {nombre_tabla} ha sido creada exitosamente.")
```
